## 7. Investigação de valores nulos
Agora vamos investigar valores ausentes.

Nem todo nulo é erro.

Um valor nulo pode significar:

dado perdido;
campo não preenchido;
informação que não se aplica;
ausência com significado de negócio;
problema de extração ou integração dos dados.

Antes de preencher ou remover nulos, investigue.

```python
nulos = df.isna().sum().sort_values(ascending=False)
nulos
```

```python
percentual_nulos = (df.isna().mean() * 100).sort_values(ascending=False)
percentual_nulos
```

```python
resumo_nulos = pd.DataFrame({
    "qtd_nulos": nulos,
    "percentual_nulos": percentual_nulos
})

resumo_nulos[resumo_nulos["qtd_nulos"] > 0]
```

## 8. Configurando análise de nulos no ydata-profiling
Podemos pedir ao ydata-profiling para gerar diagramas de nulos.

Os principais são:

bar: barras com quantidade de nulos por coluna;
matrix: padrão de ausência por linha;
heatmap: relação entre ausências de diferentes colunas.

```python
profile_nulos = ProfileReport(
    df,
    title="Relatório com foco em valores nulos",
    explorative=True,
    missing_diagrams={
        "bar": True,
        "matrix": True,
        "heatmap": True
    }
)

profile_nulos.to_file("relatorio_nulos.html")
```

```python
profile_nulos.to_notebook_iframe()
```

Limpar dados é a parte operacional. É o momento em que usamos o código para arrumar a casa: rodamos comandos para apagar linhas duplicadas, preenchemos células que vieram em branco, corrigimos erros óbvios de digitação (como diárias negativas) e removemos o que atrapalha a leitura do algoritmo. É uma atividade de engenharia, onde seguimos regras para deixar a base padronizada.

Entender o dado significa perceber que um registro "estranho" (como uma reserva feita com mais de um ano de antecedência) não é necessariamente um erro de sistema, mas sim o comportamento real de uma agência de turismo fechando um grande pacote de viagens.
